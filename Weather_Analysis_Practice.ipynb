```python
#Generate the numnbers into the Latitudes Library
def latitudes(size):
    latitudes = []
    x = 0
    while x < (size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
        x += 1
    return latitudes
# Call the function with 1500. 
#Call the function with 1500
%timeit latitudes(1500)
```

```python
print(latitudes(1500))
```
